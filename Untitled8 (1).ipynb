##Проверка на нормальность распределения экранного времени

###В рабочие дни

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Замените 'filename.xlsx' на путь к вашему файлу XLSX
df = pd.read_excel('data.xlsx', skiprows=1)

# Преобразуйте третий столбец в числовой тип данных сразу после чтения
data = pd.to_numeric(df.iloc[:, 7], errors='coerce')

# Убедитесь, что в данных нет пропущенных значений
data = data.dropna()


# Постройте гистограмму распределения данных
plt.hist(data, bins=10, edgecolor='black')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Распределение экранного времени в рабочие дни')
plt.grid(True)
plt.show()

# Выполните статистические тесты на нормальность
shapiro_test = stats.shapiro(data)
print("Shapiro-Wilk Test Statistic:", shapiro_test.statistic)
print("Shapiro-Wilk Test p-value:", shapiro_test.pvalue) #Чем меньше p-value, тем сильнее доказательства против нулевой гипотезы о нормальности распределения данных. Если p-value меньше заданного уровня значимости (обычно 0.05), то нулевая гипотеза отвергается в пользу альтернативной гипотезы о не нормальном распределении данных.

anderson_test = stats.anderson(data)
print("Anderson-Darling Test Statistic:", anderson_test.statistic) #Если значение статистики теста меньше критических значений, можно сделать вывод, что данные происходят из нормального распределения с заданным уровнем значимости.
print("Anderson-Darling Test Critical Values:", anderson_test.critical_values)

dagostino_test = stats.normaltest(data)
print("D'Agostino's K-squared Test Statistic:", dagostino_test.statistic)
print("D'Agostino's K-squared Test p-value:", dagostino_test.pvalue)#Для вычисления p-value в этом тесте используется статистика теста, которая базируется на асимметрии и эксцессе данных

ks_test = stats.kstest(data, 'norm')
print("Kolmogorov-Smirnov Test Statistic:", ks_test.statistic)
print("Kolmogorov-Smirnov Test p-value:", ks_test.pvalue)#Если p-value низкое (обычно меньше выбранного уровня значимости, например, 0.05), мы отвергаем нулевую гипотезу и заключаем, что выборки взяты из разных распределений.

```
